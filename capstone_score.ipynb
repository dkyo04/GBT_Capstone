```python
#ì´ìƒì¹˜ ë¶„ì„ 

def get_outlier(df=None, column=None, weight=1.5):
  # target ê°’ê³¼ ìƒê´€ê´€ê³„ê°€ ë†’ì€ ì—´ì„ ìš°ì„ ì ìœ¼ë¡œ ì§„í–‰
  quantile_25 = np.percentile(df[column].values, 25)
  quantile_75 = np.percentile(df[column].values, 75)

  IQR = quantile_75 - quantile_25
  IQR_weight = IQR*weight
  
  lowest = quantile_25 - IQR_weight
  highest = quantile_75 + IQR_weight
  
  # outlier_idx = df[column][ (df[column] < lowest) | (df[column] > highest) ].index
  return lowest,highest

# í•¨ìˆ˜ ì‚¬ìš©í•´ì„œ ì´ìƒì¹˜ ê°’ ì‚­ì œ
low, high= get_outlier(df=mara_df, column='2', weight=1.5)
```

```python
new_score=[]
for row in hoho['2']:
  if row >high:
    new_score.append(high)
  elif row < low:
    new_score.append(low)
  else:
    new_score.append(row)

hoho['score']=new_score
```

```python
for df in [DF_panda,DF_hj,DF_hk]:
  new_score=[]
  for row in df['2']:
    if row >high:
      new_score.append(high)
    elif row < low:
      new_score.append(low)
    else:
      new_score.append(row)

  df['score']=new_score
```
